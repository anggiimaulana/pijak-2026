## Import Library

```python
from google_play_scraper import app, Sort, reviews_all
import pandas as pd
import csv
```

## Scraping dan Simpan ke Format .csv

```python
# Scraping ulasan
scrapreview = reviews_all(
    'blibli.mobile.commerce',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=1000
)

# Simpan ke CSV langsung
with open('ulasan_aplikasi_blibli.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])
    for review in scrapreview:
        writer.writerow([review['content']])

# Buat DataFrame
app_reviews_df = pd.DataFrame(scrapreview)
app_reviews_df.to_csv('ulasan_aplikasi_blibli.csv', index=False)

print(f"Total data: {len(app_reviews_df)}")
print(f"Jumlah kolom: {len(app_reviews_df.columns)}")
app_reviews_df.head()
```
